## Import Libraries

```python
from src.utils import load, make_noise,show_images,plot_history
from src.model import AutoencoderModel
from src.train import train
import tensorflow as tf
```

```python
x_train, x_test = load()
x_train_noisy, x_test_noisy = make_noise(x_train, x_test)
```

```python
input_shape = (28, 28, 1)  # Example input shape; modify as needed
autoencoder = AutoencoderModel(input_shape).get_model()
```

## Training

```python
history = train(autoencoder, x_train_noisy, x_train, x_test_noisy, x_test,epochs=100, model_save_path='best_autoencoder.h5')
```

## Loading the saved model

```python
model = tf.keras.models.load_model('Saved Model/best_autoencoder.h5')
```

```python
denoised_images = model.predict(x_test_noisy)
```

## Results

```python
show_images(x_test_noisy, denoised_images, x_test)
```
